```python
import numpy as np 
import matplotlib.pyplot as plt 

plt.axis([-10, 10, -10, 10])
plt.grid()
plt.show()
```

```python
# Θα θεωρήσω ένα μικρό αριθμό από σημεία.
# N = 5, και θα τα τοποθετήσω σε έναν πίνακα 3xN : Κάθε στήλη αντιστοιχεί σε ένα σημείο.
X = np.array([  [0, 0, 2, 2, 1],
                [0, 2, 0, 2, 3],
                [1, 1, 1, 1, 1] ])
print(X)
plt.scatter(X[0,:], X[1,:])
plt.axis([-10, 10, -10, 10])
plt.grid()
plt.show()
```

```python
T = np.array([ [0.5, 0,   3],
               [0,   0.5, 2],
               [0,   0,   1] ])
Xout = T @ X  # @ = πολλαπλασιασμός πινάκων
print(Xout)
plt.scatter(X[0,:], X[1,:])
plt.scatter(Xout[0,:], Xout[1,:], c='r')
plt.axis([-10, 10, -10, 10])
plt.grid()
plt.show()
```

```python
points_input = np.zeros([3, 100])
points_input[2, :] = 1
for i in range(10):
    for j in range(10):
        pos = i*10 + j
        points_input[0, pos] = i
        points_input[1, pos] = j
plt.scatter(points_input[0, :], points_input[1, :])
plt.axis([-12, 12, -12, 12])
plt.grid()
plt.show()
```

```python
theta = 1
Trot = np.array([ [np.cos(theta), -np.sin(theta), 0],
                  [np.sin(theta),  np.cos(theta), 0],
                  [0,             0,              1]
                ])
points_out = Trot @ points_input
plt.scatter(points_input[0,:], points_input[1,:])
plt.scatter(points_out[0,:], points_out[1,:], c='r')
plt.axis([-12, 12, -12, 12])
plt.grid()
plt.show()
```

```python
tranX = 3
tranY = 4
Ttranslate = np.array([ [1,  0,  tranX],
                        [0, 1,  tranY],
                        [0, 0,   1] ])
Tsynthesis = Trot @ Ttranslate
Tsynthesis2 = Ttranslate @ Trot


points_out2 = Tsynthesis2 @ points_input
plt.scatter(points_input[0,:], points_input[1,:])
plt.scatter(points_out2[0,:], points_out2[1,:], c='r')
plt.axis([-12, 12, -12, 12])
plt.grid()
plt.show()
```

```python
tranX = -5
tranY = -5
Ttranslate = np.array([ [1,  0,  tranX],
                        [0, 1,  tranY],
                        [0, 0,   1] ])
Tsynthesis3 = np.linalg.inv(Ttranslate) @ Trot @ Ttranslate
points_out2 = Tsynthesis3 @ points_input
plt.scatter(points_input[0,:], points_input[1,:])
plt.scatter(points_out2[0,:], points_out2[1,:], c='r')
plt.axis([-12, 12, -12, 12])
plt.grid()
plt.show()
```
